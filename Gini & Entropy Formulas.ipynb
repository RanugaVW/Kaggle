```python
import math
```

```python
def gini(samples):
    samples_sum = sum(samples)
    score = 0
    probas  = []
    for sample in samples:
        proba = sample / samples_sum
        probas.append(proba)
    for i in probas:
        score += i * (1-i)
    return score
```

```python
def entropy(samples):
    samples_sum = sum(samples)
    score = 0
    probas  = []
    for sample in samples:
        proba = sample / samples_sum
        probas.append(proba)
    for i in probas:
        if i > 0:
            score += i * math.log(i)
    return -score
```

```python
samples = [100, 75, 125]
gini(samples), entropy(samples)
```
